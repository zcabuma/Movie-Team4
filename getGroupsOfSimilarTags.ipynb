```


def letter_cmp(a, b):
    word1 = nlp.vocab[str(a[0])]
    word2 = nlp.vocab[str(b[0])]
    wordActual = nlp.vocab[str(b[1])]

    simScore1 = word1.similarity(wordActual)
    simScore2 = word2.similarity(wordActual)


    if simScore1 > simScore2:
        return -1
    else:
        return 1

from functools import cmp_to_key
letter_cmp_key = cmp_to_key(letter_cmp)

# tagList = ['funny', 'joke', 'laugh', 'sad']


visited = set();
index = 0
result = pd.DataFrame(columns=['tag', 'group'])

for word in processedTagList:
  if word not in visited:
    currentList =  []
    for e in processedTagList:
      word1 = nlp.vocab[str(e)]
      word2 = nlp.vocab[str(word)]
      simScore1 = word1.similarity(word2)
      if simScore1 > 0.5 and e not in visited:
        currentList.append((e, word))
        

    currentList.sort(key=letter_cmp_key)
    visited.add(word)
    print(currentList[:3])
    for element in currentList[:3]:
      entry = {'tag': element[0], 'group': index}
      print(entry)
      visited.add(element[0])
      result = result.append(entry, ignore_index=True)
    index = index + 1


result.to_csv("result.csv");
```
